## Make features using 100 and 200

```python
import scipy.ndimage


def data_gen(file_loc, batch_size, return_binary=False, magnification=None, im_size=None, 
    square_rot_p=.3, translate=0, flips=False, rotate=False, blur=False, zoom=1, ):
    # square_rot_p is the prob of using a 90x rotation, otherwise sample from 360. Possibly not useful
    # translate is maximum number of pixels to translate by. Make it close the doctor's variance in annotation

    cancer_classes=['B_A', 'B_F', 'B_PT', 'B_TA', 'M_DC', 'M_LC', 'M_MC', 'M_PC']

    square_rot_p = int(square_rot_p)
    translate = int(translate)

    all_files=glob.glob(os.path.join(file_loc, '*'))
    if int(magnification)!=0:
        all_files = [loc for loc in all_files if loc.rsplit('/', 1)[1].rsplit('-', 1)[0].rsplit('-', 1)[1] == str(magnification)]

    while 1:
        random.shuffle(all_files) # randomize after every epoch
        num_batches = int(np.floor(len(all_files)/batch_size))

        for batch in range(num_batches):
            x=[]
            y=[]
            batch_files = all_files[batch_size*batch:batch_size*(batch+1)]
            for image_loc in batch_files:
                image = Image.open(image_loc)

                # APPLY AUGMENTATION:
                # flips
                if flips:
                    flip_vert = random.randint(0, 1)
                    flip_hor = random.randint(0, 1)
                    if flip_vert:
                        image = image.transpose(Image.FLIP_TOP_BOTTOM)
                    if flip_hor:
                        image = image.transpose(Image.FLIP_LEFT_RIGHT)

                # rotation
                if rotate:
                    square_rot =  bool((np.random.uniform(0, 1, 1)<square_rot_p))
                    if square_rot:  # maybe this is dumb, but it cant hurt
                        angle = random.randint(0, 4)
                        if(angle ==0):
                            image = image.transpose(Image.ROTATE_90)
                        elif(angle ==1):
                            image = image.transpose(Image.ROTATE_180)
                        elif(angle ==2):
                            image = image.transpose(Image.ROTATE_270)
                    else:
                        angle = np.random.uniform(0, 360,1)
                        image=image.rotate(angle)

                #blur
                if blur:
                    if random.randint(0, 1):
                        radius = random.randint(1, 4)
                        image = image.filter(ImageFilter.GaussianBlur(radius=radius))

                if(im_size != 0):
                    image_shape = (im_size, im_size)
                    image = image.resize(image_shape)

                # translate
                ts_sz_row = randint(-1*translate, translate)
                ts_sz_col = randint(-1*translate, translate)

                image = image.transform(image.size, Image.AFFINE, (1, 0, ts_sz_row, 0, 1, ts_sz_col))

                image = np.reshape(np.array(image.getdata()), (im_size, im_size, 3))
                image = image/255.0
                if (zoom!=1):
                    image = scipy.ndimage.zoom(image, 2)
                    zoomed_img = np.zeros(image.shape*2)
                    x_offset = image.shape[0]/2
                    y_offset = image.shape[1]/2
                    zoomed_img[int(x_offset):int(image.shape[0]+x_offset),int(y_offset):int(image.shape[1]+y_offset)] = image
                    image = zoomed_img 

                label = image_loc.rsplit('/', 1)[1].split('_', 1)[1].split('-', 1)[0]
                if return_binary:
                    y_temp = int(label in cancer_classes[4:])
                    y_temp = np.eye(2)[y_temp]
                else :
                    y_temp = cancer_classes.index(label)
                    y_temp = np.eye(8)[y_temp]
                
                x.append(image)
                y.append(y_temp)
            x=np.array(x)
            y=np.array(y)
            yield (x, y)
            
```

## Test models on these features 

```python
# load the train data
features = np.load('/Users/rb/Documents/waterloo/projects/breakHis/features/feat_vgg_100_allaug2_16.npy')
print(features.shape)

y = features[:,:8]
x_train = features[:,8:]

# Make y-values indicating benign vs. malignant
y_bin_train = np.zeros((y.shape[0],2))
for index, row in enumerate(y):
    if(np.sum(row[:4])>0):
        y_bin_train[index, 0] = 1
    else:
        y_bin_train[index, 1] = 1
        

# load the test data
features = np.load('/Users/rb/Documents/waterloo/projects/breakHis/features/feat_vgg_100_test.npy')

y = features[:,:8]
x_test = features[:,8:]

# Make y-values indicating benign vs. malignant
y_bin_test = np.zeros((y.shape[0],2))
for index, row in enumerate(y):
    if(np.sum(row[:4])>0):
        y_bin_test[index, 0] = 1
    else:
        y_bin_test[index, 1] = 1

        
# get rid of hot-one
y_bin_train = np.argmax(y_bin_train, axis=1)
y_bin_test = np.argmax(y_bin_test, axis=1)
```

## Two class:
* no rebalancing for 2 classes

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

clf = RandomForestClassifier(n_estimators=50)
clf.fit(x_train, y_bin_train)
y_pred = clf.predict(x_train)
train_acc = accuracy_score(y_bin_train, y_pred)

# check test acc
y_pred_test = clf.predict(x_test)
test_acc=accuracy_score(y_bin_test, y_pred_test)

print('Train acc: ', train_acc)
print('Test acc: ', test_acc)
```

```python
# XGBoost
import xgboost as xgb
from sklearn.metrics import accuracy_score

clf = xgb.XGBRegressor()
clf.fit(x_train, y_bin_train)
y_pred = clf.predict(x_train)
train_acc = accuracy_score(y_bin_train, np.round(y_pred))
# check test acc
y_pred = clf.predict(x_test)
test_acc=accuracy_score(y_bin_test, np.round(y_pred))

y_pred = clf.predict(x_test)
test_acc=accuracy_score(y_bin_test, np.round(y_pred))
print('XGB Accuracy train: ', train_acc)
print('XGB Accuracy test: ', test_acc)
```

```python
# Linear SVM
from sklearn import svm

clf = svm.SVC(kernel='linear')
clf.fit(x_train, y_bin_train)
y_pred = clf.predict(x_train)
train_acc = accuracy_score(y_bin_train, y_pred)

# check test acc
y_pred = clf.predict(x_test)
test_acc=accuracy_score(y_bin_test, y_pred)

print('Train acc: ', train_acc)
print('Test acc: ', test_acc)
```

```python
# Polynomial SVM (3rd degree)
from sklearn import svm

clf = svm.SVC(kernel='poly')
clf.fit(x_train, y_bin_train)
y_pred = clf.predict(x_train)
train_acc = accuracy_score(y_bin_train, y_pred)

# check test acc
y_pred = clf.predict(x_test)
test_acc=accuracy_score(y_bin_test, y_pred)

print('Train acc: ', train_acc)
print('Test acc: ', test_acc)
```
